```python
import torch 
import numpy as np 

```

```python
#train data and targets 

train_data = torch.tensor(
    [
        [0, 0],
        [0, 1],
        [1, 0],
        [1, 1]], dtype=torch.float32)

target_xor = torch.tensor(
    [
        [0],
        [1],
        [1],
        [0]], dtype=torch.float32)

```

# Linear regression model

```python
# linear regression model 

class LinearRegression(torch.nn.Module):
    
    def __init__(self, input_size, output_size):
        super(LinearRegression, self).__init__()
        self.linear = torch.nn.Linear(input_size, output_size)
        
    def forward(self, x):
    
        return self.linear(x)
```

```python
#train the linear model 
linear_model= LinearRegression(input_size=2, output_size=1)
criterion= torch.nn.MSELoss()
optimizer= torch.optim.SGD(linear_model.parameters(), lr=0.1)

for epoch in range(1000):
    #forward 
    outputs=linear_model(train_data)
    loss=criterion(outputs, target_xor)
    
    #backward and optimization 
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print("epoch {}, loss function {}".format(epoch, loss.item()))

# there is no convergence, which shows that the linear model failed to learn such mapping
```

```python
# testing the model 
linear_model.eval()
with torch.no_grad():
    linear_outputs=linear_model(train_data)
    linear_predicitions= (linear_outputs > 0.5).float()
    linear_accuracy=(linear_predicitions==target_xor).float().mean().item()
    print("linear model accuracy:{:.2f}".format(linear_accuracy))
```

# 2-layer NN model


```python
# define 2 layer NN

class NeuralNetwork(torch.nn.Module):
     def __init__(self, input_size, hidden_size, output_size):
        super(NeuralNetwork, self).__init__()
        self.linear1 = torch.nn.Linear(input_size, hidden_size)
        self.activation= torch.nn.ReLU()
        self.linear2 = torch.nn.Linear(hidden_size, output_size)
     def forward(self, x):
        x=self.linear1(x)
        x=self.activation(x)
        x=self.linear2(x)
        return x
    
    
```

```python
#train and optimization 

neural_model=NeuralNetwork(input_size=2, hidden_size=4, output_size=1)
criterion=torch.nn.MSELoss()
optimizer=torch.optim.SGD(neural_model.parameters(), lr=0.1)

for epoch in range(10000):
    #forward 
    outputs=neural_model(train_data)
    loss=criterion(outputs, target_xor)
    
    #backward and optimization 
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print("epoch {}, loss function {}".format(epoch, loss.item()))
```

```python
# testing the model  
neural_model.eval()
with torch.no_grad():
    neural_outputs=neural_model(train_data)
    neural_predicitions= (neural_outputs > 0.5).float()
    neural_accuracy=(neural_predicitions==target_xor).float().mean().item()
    print("neural network accuracy:{:.2f}".format(neural_accuracy))

```
